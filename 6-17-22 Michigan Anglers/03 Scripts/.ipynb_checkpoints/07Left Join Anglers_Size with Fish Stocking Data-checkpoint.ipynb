```python
#import csv master angler & sizes dataframe
df_anglers = pd.read_csv(r'C:\Users\Selena\Desktop\6-17-22 Michigan Anglers\02 Data\Prepared\anglers_size_combined.csv', index_col=False)
path = r'C:\Users\Selena\Desktop\6-17-22 Michigan Anglers'
df_anglers = pd.read_csv(os.path.join(path, '02 Data', 'Prepared', 'anglers_size_combined.csv'), index_col=False)
```

# Importing fish stocking dataframe

```python
#import csv fish stocking dataframe
df_stocking = pd.read_csv(r'C:\Users\Selena\Desktop\6-17-22 Michigan Anglers\02 Data\Original\3fish stocking data.csv')
path = r'C:\Users\Selena\Desktop\6-17-22 Michigan Anglers'
df_stocking = pd.read_csv(os.path.join(path, '02 Data', 'Original', '3fish stocking data.csv'))
```

```python
df_stocking.info()
```

# Doing a Left Merge

```python
#need to do a left join with anglers on left and stocking on right using species as the way to merge
df_merged = df_anglers.merge(df_stocking, on = (['year','species']), how = 'left', indicator = True)
```

# Checking Results

```python
df_merged.head()
```

```python
#checking the shape
df_merged.shape
```

```python
# checking the rows merged on species
df_merged['_merge'].value_counts()
```

```python
#seeing where data is missing
df_merged.isnull().sum()
```

#this can make sense... not all species may have been stocked

#I need to see if the species are probably not stocked

```python
df_null = df_merged[df_merged.isna().any(axis=1)]
```

```python
df_null.head()
```

#I did a quick search and the species with NA for number_stocked were not stocked. I will replace NA with 0

# Filling null values with zero because the species of fish were not stocked that year

```python
df_merged['number_stocked'] = df_merged['number_stocked'].fillna(0)
```

```python
#seeing where data is missing
df_merged.isnull().sum()
```

#no null values

```python
#need to drop the _merged column
#need to remove _merge
df_merged = df_merged.drop('_merge', 1)
```
