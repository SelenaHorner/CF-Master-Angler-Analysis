# Getting ready to create a visual of the various species in the Great Lakes

```python
#only looking at one year
df_1999 = df_anglers[df_anglers['year']==1999]
```

```python
#only looking at chinook salmon
df_1999_chinook_salmon = df_1999[df_anglers['species']=='chinook salmon']
```

```python
#checking change
df_1999_chinook_salmon.head()
```

```python
# That's just in case you want to look at the JSON file contents here too:

f = open(r'C:\Users\Selena\Desktop\6-17-22 Michigan Anglers\02 Data\Original\6great lakes.geojson',)
  
# returns JSON object asa dictionary
data = json.load(f)
  
# Iterating through the json list
for i in data['features']:
    print(i)
```

```python
df_1999_chinook_salmon.info()
```

```python
# Create a data frame with just the lakes and the chinnook salmon length

data_to_plot = df_1999_chinook_salmon[['merge','length in']]
data_to_plot.head()
```

```python
# Setup a folium map at a high-level zoom
map = folium.Map(location = [100, 0], zoom_start = 1.5)

# Choropleth maps bind Pandas Data Frames and json geometries.This allows us to quickly visualize data combinations
folium.Choropleth(
    geo_data = gl_geo, 
    data = data_to_plot,
    columns = ['merge', 'length in'],
    key_on = 'properties.merge', # this part is very important - check your json file to see where the KEY is located
    fill_color = 'BuGn', fill_opacity=0.6, line_opacity=0.1,
    legend_name = "length").add_to(map)
folium.LayerControl().add_to(map)

map
```
